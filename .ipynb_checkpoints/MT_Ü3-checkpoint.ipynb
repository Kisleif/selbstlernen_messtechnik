## Ü3.1 Dichtemessung einer Flüssigkeit: Messabweichung <a id="Sec-31"></a> [](#Ü31)

Die Dichte eines flüssigen Mediums kann unter Ausnutzung des Archimedischen Prinzips ermittelt werden. Hiernach taucht ein Körper so weit in eine Flüssigkeit ein, bis die Gewichtskraft der verdrängten Flüssigkeit der Gewichtskraft des eingetauchten Körpers entspricht. Je kleiner also die Dichte der zu untersuchenden Flüssigkeit ist, desto tiefer wird ein schwimmender Prüfkörper in diese eintauchen.

Für entsprechende Messungen verwendete Prüfkörper heißen Senkspindel oder Aräometer und bestehen vereinfacht aus einem zylindrischen Glaskörper, wie in der Abbildung skizziert:

```python
Image("u3_abb1.png")
```

Damit der Prüfkörper in der Flüssigkeit in aufrechter Lage stabil schwimmt, ist er in seinem unteren Teil mit einem meist aus Blei bestehenden Gewicht beschwert. Der zylindrische Glaskörper trägt eine Skala, an welcher die Eintauchtiefe des Prüfkörpers abgelesen werden kann.

Im vorliegenden Fall weise der Glaskörper die Masse 􏲵􏲶􏲷􏲸􏲹$m_\mathrm{Glas}$ auf und sei mit einem zusätzlichen Bleigewicht der Masse $m_\mathrm{Blei}$􏲵􏲺􏲷􏲻􏲼 beschwert. Die Eintauchtiefe, gemessen von der Unterkante des zylindrischen Prüfkörpers bis zum Spiegel der Flüssigkeit, wird von der Skala angezeigt und sei mit $h$ bezeichnet. Der zylindrische Prüfkörper weise über die gesamte Länge den Durchmesser $d$ 􏲽 auf. Die Dichte $\rho$􏲾 des zu untersuchenden Mediums ist dann näherungsweise durch folgenden formelmäßigen Zusammenhang definiert:

$$\rho = \frac{m_\mathrm{Blei} + m_\mathrm{Glas}}{\frac{1}{4} \pi d^2 h }$$

Im Folgenden soll die Dichte $\rho$􏲾 auf der Grundlage von Messergebnissen für die Größen 􏲵􏲶􏲷􏲸􏲹$m_\mathrm{Glas}$, $m_\mathrm{Blei}$􏲵􏲺􏲷􏲻􏲼, $d$􏲽 und $h$ einschließlich der wahrscheinlichen Abweichungsgrenzen ermittelt werden.

Die Massen $m_\mathrm{Glas}$􏲵􏲶􏲷􏲸􏲹 und 􏲵􏲺􏲷􏲻􏲼$m_\mathrm{Blei}$ werden in einer gemeinsamen Wägung mittels einer elektronischen Präzisionswaage ermittelt. Die von der Waage angezeigte gemeinsame Masse von Glaskörper und Bleigewicht beträgt 􏳁􏲵􏲶􏲷􏲸 $M = (m_\mathrm{Glas}+m_\mathrm{Blei}) 􏲵􏲺􏲷􏲻􏲼 = 14\,\mathrm g$. Die relative Unsicherheit der Waage wird vom Hersteller mit 0,5% des Anzeigewertes bei einer Aussagewahrscheinlichkeit von 􏳃$P = 98\%$ angegeben.

```python
M = 14e-3 # Gesamtmasse in kg
u_M_rel = 0.005 # relative Unsicherheit
u_M = u_M_rel*M # absolute Unsicherheit
print('Die absolute Unsicherheit der Gesamtmasse für P = 98% beträgt u_M =', u_M, 'kg')
```

### Berechnung des Durchmessers <a id="Ü31.a"> </a>

Der Durchmesser 􏲽$d$ des Glaszylinders wurde in $n_d = 25$ Wiederholungen mittels einer Bügelmessschraube gemessen. Das vollständige Messergebnis des Durchmessers beträgt $d􏲽 = 12\,\mathrm{mm} \pm 0,011\,\mathrm{mm}$ bei einer Aussagewahrscheinlichkeit von 􏳃$P = 90\%$. Allgemein gilt für die Umrechnung zwischen zwei Aussagewahrscheinlichkeiten folgender Zusammenhang für die Unsicherheiten:

$$ u_{\alpha 1} = u_{\alpha 2} \cdot \frac{t_{n-1; 1-\alpha 1 / 2}}{t_{n-1; 1-\alpha 2 / 2}}$$

```python
d = 12e-3 # Durchmesser in m
P2 = 0.9 # gegebene Aussagewahrscheinlichkeit
P1 = 0.98 # gewünschte Aussagewahrscheinlichkeit
n_d = 25 # Anzahl der Wiederholungen
alpha1 = 1-P1
alpha2 = 1-P2
print('In der Tabelle für die Quantile der Student-t Verteilung müssen folgende Parameter berücksichtigt werden:')
print('t1 = t_', n_d-1, ';',1-alpha1/2)
print('t2 = t_', n_d-1, ';',1-alpha2/2)
```

```python
Image("u3_abb2.png")
```

```python
t1 = 2.492
t2 = 1.711
u_alpha2 = 0.011e-3 # Unsicherheit in m für P = 90%
u_alpha1 = u_alpha2 * t1 / t2 # Berechnunge der Unisicherheit in m für P = 98%
u_d = u_alpha1
print('Die absolute Unsicherheit des Durchmessers in m beträgt für P = 98%: u_d = ', u_d, 'm')
```

### Berechnung der Eintauchtiefe <a id="Ü31.b"> </a>

Die Eintauchtiefe h wird in insgesamt $n_h􏳄􏳆 = 7$ Wiederholungen von der Skala abgelesen. Dabei werden die in der Tabelle zusammengefassten Einzelmesswerte ermittelt: 

```python
df_raw = pd.read_csv('u3_data.csv')
df_raw.head(8)
```

Der Mittelwert wird analytisch mittels

$$\overline h = \frac{1}{n_h} \sum_{i=1}^7 h_i$$

berechnet.

```python
df = df_raw*1e-3 # in m
h = df.mean()
print('Der Mittelwert ist: ', h.values, 'm')
```

Der Standardabweichung berechnet sich analytisch mittels:

$${\overline h} = \sqrt{\frac{1}{n_h - 1} \sum_{i=1}^7 (h_i-\overline h)^2}$$

```python
s_h = df.std()
print('Die Standardabweichung s_h ist:  ', s_h.values, 'm')
```

Der Vertrauensbereich der Messgröße berechnet sich über die Student-t Verteilung:

$$ u_h = \frac{s_h}{\sqrt{n_h}} \cdot t_{6;0,99} $$

```python
t_6_099 = 3.143
u_h = s_h / np.sqrt(7) * t_6_099
print('Die absolute Unsicherheit der Eintauchtiefe h in m beträgt für P = 98%: u_h = ', u_h.values, 'm')
```

### Berechnung der Dichte <a id="Ü31.c"> </a>

Berechnung des Mittelwertes für die gesuchte Größe $\rho$ ermittelt sich mittels

$$ \overline \rho = \frac{\overline M}{\frac{1}{4}\pi \overline d^2 \overline h}.$$

```python
rho = M/(0.25 * np.pi * d**2 * h.values)
print('Mittelwert der Dichte: ', rho, 'kg / m^3')
```

Für die Fehlerfortpflanzung nach Gauß wird folgende Formel verwendet:
    
$$ u_\rho = \sqrt{\left(\frac{\partial \rho}{\partial M} \cdot u_M\right)^2 + \left(\frac{\partial \rho}{\partial d} \cdot u_d\right)^2 + \left(\frac{\partial \rho}{\partial h} \cdot u_h\right)^2} $$

wobei für die partiellen Ableitungen folgendes gilt:

$$ \frac{\partial \rho}{\partial M} = \frac{1}{\frac{1}{4}\pi \overline d^2 \overline h}$$

$$ \frac{\partial \rho}{\partial d} = -\frac{2M}{\frac{1}{4}\pi \overline d^3 \overline h}$$

$$ \frac{\partial \rho}{\partial h} = -\frac{M}{\frac{1}{4}\pi \overline d^2 \overline h^2}$$

```python
Drho_Dm = 1/(0.25 * np.pi * d**2 * h)
Drho_Dd = -2*M /(0.25 * np.pi * d**3 * h)
Drho_Dh = -M /(0.25 * np.pi * d**2 * h**2)
u_rho = np.sqrt((Drho_Dm * u_M)**2 + (Drho_Dd * u_d)**2 + (Drho_Dh * u_h)**2)
print('Die Unsicherheit der Dichte beträgt: ', u_rho.values, 'kg/m^3')
```

Das vollständige Messergebnis lautet:

```python
print(rho, 'kg/m^3 +-', u_rho.values, 'kg/m^3 für P = 98%')
```

## Ü3.2 Quantisierungsabweichung infolge der Digitalisierung  <a id="Sec-32"></a> [](#Ü32)

### Relative Abweichung <a id="Ü32.a"> </a>
Die relative Abweichung infolge der Quantisierung soll kleiner als 0,02\% bleiben. 
Damit dies gewährleistet sein kann, muss die Bit-Auflösung des zu verwendenden AD-Wandlers bestimmt werden:

$$A_\mathrm{rel} = \frac{1}{2^n-1} < 0,02\% = 0,0002$$

Umstellen der Formel nach $n$ liefert:

$$2^n -1 > \frac{1}{0,0002} $$

$$2^n >  \frac{1}{0,0002} + 1 = 5000 + 1 = 5001$$

$$\lg 2^n = n \cdot \lg 2 > \lg 5001 $$

$$n = \frac{\lg 5001}{\lg 2}$$

```python
n = np.log(5001)/np.log(2)
print(n, 'Bit')
```

Es ist ein AD-Wandler mit einer Auflösung von 􏰌>13 Bit erforderlich. In der Praxis wird in diesem Fall wohl aus Verfügbarkeitsgründen vorrangig ein AD-Wandler mit einer Auflösung von 14 Bit verwendet werden.

### Digitalisierung eines Spannungssignals <a id="Ü32.b"> </a>

AD-Wandler werden häufig mit einem Full-Scale-Wert FS = 10 V realisiert. 
Der kleinste Spannungswert, der noch ein Umschalten des LSB bewirkt wird für einen 10, 14 und 16-Bit AD-Wandler mittels

$$A_\mathrm{rel} = \frac{A_\mathrm{max}}{FS} = \frac{1 \mathrm{LSB}}{FS} = \frac{1 \mathrm{LSB}}{(2^n-1)1 \mathrm{LSB}}$$

$$ \rightarrow 1 \mathrm{LSB} = \frac{FS}{2^n-1}$$

berechnet. 

```python
print('10 Bit: ', 10 / (2**(10)-1), 'V = ', 10 / (2**(10)-1)*1e3, 'mV')
print('14 Bit: ', 10 / (2**(14)-1), 'V = ', 10 / (2**(14)-1)*1e3, 'mV')
print('16 Bit: ', 10 / (2**(16)-1), 'V = ', 10 / (2**(16)-1)*1e3, 'mV')
```

## Ü3.3 Hall-Konstante: Lineare Regression  <a id="Sec-33"></a> [](#Ü33)

Um die Hall-Konstante $A_\mathrm H$ eines neuen Werkstoffs zu bestimmen, haben Sie eine Messreihe durchgeführt, bei welcher in bestimmten Arbeitspunkten jeweils Strom und Spannung an einem Hall-Element gemessen wurden. Unter Berücksichtigung der relevanten Konstanten – magnetische Flussdichte und Dicke des Hall-Elements – erhalten sie die in nachfolgender Tabelle zusammengefassten $x-y$-Wertepaare:

```python
df_hall = pd.read_csv('u3_data_hall.csv', header=None,index_col=None, names=['x/kVCm^3','y/V'])
df_hall.head(8)
```

Die gesuchte Hall-Konstante $A_\mathrm H$ ergibt sich als Regressionskoeffizient aus obigen $x-y$-Wertepaaren.

```python
x = df_hall['x/kVCm^3']
y = df_hall['y/V']
plt.scatter(x,y)
plt.xlabel('x/kVCm^3')
plt.ylabel('y/V')
```

### Korrelationskoeffizient <a id="Ü33.a"> </a>

Der Korrelationskoeffizient berechnet sich aus

$$r = \frac{\mathrm{cov}_{xy}}{s_x \cdot s_y} = \frac{\sum_{i=1}^N (x_i-\overline x)\cdot (y_i-\overline y)}{\sqrt{\sum_{i=1}^N (x_i-\overline x)^2} \cdot \sqrt{ \sum_{i=1}^N (y_i-\overline y)^2}}$$

```python
x_mw = np.mean(x)                     # Berechnung des Mittelwertes von x
y_mw = np.mean(y)                     # Berechnung des Mittelwertes von y
s_x = np.sqrt(np.sum( (x-x_mw)**2))   # Berechnung der Standardabweichugn von x
s_y = np.sqrt(np.sum( (y-y_mw)**2))   # Berechnung der Standardabweichugn von y
cov_xy = np.sum( (x-x_mw)*(y-y_mw) )  # Berechnung der Kovarianz 
r = cov_xy / (s_x*s_y)                # Berechnung der Korrelationskoeffizienten
print('Der Korrelationskoeffizient beträgt r = ',r)
```

### Lineare Regression <a id="Ü33.b"> </a>

Die lineare Regression

$$y = mx +b$$

lässt sich die Steigung $m$ mittels

$$ m = \frac{N \cdot \sum_{i=1}^N (x_i y_i) - \sum_{i=1}^N x_i \sum_{i=1}^N y_i}{N \cdot x_i^2 - \left( \sum_{i=1}^N x_i\right)^2}$$

berechnen und die Achsenabschnitt mittels

$$ b = \frac{\sum_{i=1}^N y_i - m \sum_{i=1}^N x_i}{N}$$

wobei $N$ die Anzahl der Messwerte ist.

```python
m = (len(x) * np.sum(x*y) - np.sum(x) * np.sum(y)) / (len(x)*np.sum(x*x) - np.sum(x) ** 2)
b = (np.sum(y) - m *np.sum(x)) / len(x)
print('Die Steigung (Hall-Konstante) ist somit m =',m,'kCm^3.' )
print(b)
```

```python
y_linreg = m*x+b
plt.scatter(x,y)
plt.xlabel('x/kVCm^3')
plt.ylabel('y/V') 
plt.plot(x,y_linreg)
```

### Lineare Regression in Python <a id="Ü33.c"> </a>

Heutzutage werden lineare Regression selten per Hand gelöst, da das Verfahren sehr aufwendig ist. Viele Programme bieten Algorithmen, welche auf den obigen Gleichungen basieren. Mithilfe dieser Algorithmen können die gesuchten Koeffizienten direkt bestimmt werden, wie beispielsweise in Python mithilfe des 'scipy.stats'-Pakets für statistische Analysen.

```python
from scipy.stats import linregress
y_py_linreg = linregress(x,y)
print('Steigung der lin. Regression: ',y_py_linreg.slope) # Steigung der Geraden y = a + b*x
print('Y-Achsenabschnitt der lin. Regression: ', y_py_linreg.intercept) # Y-Achsenabschnitt der linearen Regression
print('Korrelationskoeffizient: r =',y_py_linreg.rvalue) # Korrelationskoeffizient
#print(linreg.stderr) # Abweichung der Steigung
```

```python
plt.scatter(x,y) # Messwerte
plt.plot(x,y_linreg) # Linare Regression per Hand
plt.plot(x, y_py_linreg.intercept+y_py_linreg.slope*x) # Lineare Regression von Python
plt.xlabel('x/kVCm^3')
plt.ylabel('y/V')
```

Dies liefert natürlich auch die Möglichkeit nicht nur lineare Regressionen durchzuführen, sondern leglichen Datenansatz anhand von passenden Modellen (Fit-Funktionen) anpassen zu können. So können polynomische, periodische oder exponentielle Zusammenhänge beispielsweise einfach überprüft werden.
